```python
import pandas as pd

URL="https://drive.usercontent.google.com/download?id=1JMYqXipZpz9Y5-vyxvLEO2Y1sRBxqu-U&export=download&authuser=0"
df = pd.read_csv(URL)
assert df is not None, "DATA NOT READ FROM THE URL"
```

```python
cols_with_null = df.isnull().sum() > 0
cols_with_null["Язык.программирования"] = False
print(cols_with_null[cols_with_null == False].index)
df = df[cols_with_null[cols_with_null == False].index]
print(df.head())
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
print(df.shape)
```

```python
python_data = df[df['Язык.программирования'].str.lower() == "python"]
python_data.shape
```

```python
gr_by_posada = python_data.groupby(['Должность'])
print(gr_by_posada.count().index)
print(gr_by_posada.mean(numeric_only=True).sort_values("salary", ascending=False))
```

```python
gr_by_posada['salary'] \
  .agg(['min', 'mean', 'max']) \
  .sort_values("mean", ascending=False) \
  .astype(int)
```

```python
spec_stat = gr_by_posada['salary'].agg(['min', 'max'])
spec_stat.head()
```

```python

def fill_avg_salary(s, ref_df):
  avg = None
  try:
    sp = s.get('Должность')
    if sp:
      stat_sp = ref_df.loc[sp]
      avg=int((stat_sp["min"] + stat_sp["max"])/2)
  finally:
    s['avg']=avg
    return s


a=df.apply(fill_avg_salary, args=[spec_stat], axis=1)
a[a.columns[[0,1,2,19,20]]].head(2)
```

```python
def fill_avg(s):
  avg = None
  try:
      avg=int((s["min"] + s["max"])/2)
  finally:
    s['avg']=avg
    return s

b=spec_stat.apply(fill_avg, axis=1)
b
```
